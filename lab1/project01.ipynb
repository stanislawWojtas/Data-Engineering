```python
import pandas as pd
import numpy as np
import json
```

### Exercise 1: Column information
Loading proj1_ex01.csv into a dataframe:

```python
url = "proj1_ex01.csv"
df = pd.read_csv(url)
df
```

```python
col_names = df.columns
col_names
```

```python
missing_val = df.isnull().sum()/ df.shape[0]
missing_val = np.round(missing_val, 2)
missing_val
```

```python
types = df.dtypes
types = types.astype(str)
types = types.replace("int64", "int")
types = types.replace("float64", "float")
types = types.replace("object", "other")
types
```

```python
df_json = pd.DataFrame({"name": col_names, "missing": missing_val, "type": types})
df_json
```

```python
df_json.to_json("proj1_ex01_fields.json", orient="records", indent=4)
```

### Exercise 2: Value statistics

```python
# adding description to json file
df.describe(include='all').to_json("proj1_ex02_stats.json", indent=4)
```

### Exercise 3: Column names

```python
# Renaming column names
import re # for regular expressions
#function to rename column names
orginal_cols = df.columns
new_cols = []
for col in orginal_cols:
  col_lower = col.lower()  # Zamiana na małe litery
  col_clean = re.sub(r'[^A-Za-z0-9_ ]', '', col_lower)  # Usunięcie niedozwolonych znaków
  col_final = re.sub(' ', '_', col_clean)  # Zamiana spacji na podkreślenia
  new_cols.append(col_final)  # Dodanie przekształconej nazwy do listy
df.columns = new_cols
df
```

```python
## Saving dataframe to csv filed (without index)
df.to_csv('proj1_ex03_columns.csv', index=False)
```

### Exercise 5: Selecting rows and columns
