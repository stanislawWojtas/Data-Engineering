### Exercise 1: Load the file

```python
# Loading data from proj2_data.csv
import pandas as pd

# Open header file to check delimiter
with open('proj2_data.csv') as f:
    header = f.readline()
    if(';' in header):
      delimiter = ';'
    elif(',' in header):
      delimiter = ','
    else:
      delimiter = '|'
print("Delimiter is: ", delimiter)


df = pd.read_csv('proj2_data.csv', delimiter=delimiter)
```

```python
import re
pattern = r"-?\d+([\.,]\d+)?"
# change all decimal values with ',' to this with '.'
for col in df.columns:
  # If kolumn is type float
  if(df[col].astype(str).str.match(pattern).all()):
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace(',', '.').astype(float)
df
```

```python
# saving data frame into pickle
df.to_pickle('proj2_ex01.pkl')
```

### Exercise 2: Value scale

```python
string_to_val = {}
i = 1
with open('proj2_scale.txt') as file:
  for line in file:
    line = line.strip() # deleting whitespaces
    string_to_val[line] = i
    i += 1

# new dataFrame
df_new = df.copy()

# We search for column with values same as dictionary keys
for col in df_new.columns:
  values_set = set(df_new[col].unique())
  if(values_set.issubset(string_to_val.keys())):
    df_new[col] = df_new[col].map(string_to_val)
df_new
```

```python
# Saving to pkl file
with open('proj2_ex02.pkl', 'wb') as file:
  df_new.to_pickle(file)
```

### Exercise 3: Categories

```python
df_3 = df.copy()
for col in df_3.columns:
  values_set = set(df_3[col].unique())
  if(values_set.issubset(string_to_val.keys())):
    df_3[col] = df_3[col].astype('category')
    # adding all categories (even if not present in dataframe)
    df_3[col] = df_3[col].cat.set_categories(string_to_val.keys())

df_3.dtypes
```

```python
# Saving result to pkl file
with open ('proj2_ex03.pkl', 'wb') as file:
  df_3.to_pickle(file)
```

### Exercise 4: Number extraction

Wyjanienie regexu  $'-?\d+(?:[\.,]\d+)?'$
- $-?$ moe by minus ale nie musi
- \d+ znak liczby (moe by wiecej ni jeden)
- $(?:[\.,]\d+)?$ grupa bez zapamitania moe wystpi lub nie

```python
import re
df_4 = df.copy()
# float and int columns are already in required format so we only look at object columns
pattern = r"(-?\d+(?:[\.,]\d+)?)"
# list of columns we changed
list_of_changes_cols = []

for col in df_4.columns:
  if(df_4[col].dtype == 'object'):
    # checking if first element of the column is in format we look for
    value = df_4.iloc[0].loc[col]
    if(re.match(pattern, value)):
      # changing whole column to format we want
      df_4[col] = df_4[col].astype(str).str.replace(',', '.')
      df_4[col] = df_4[col].str.extract(pattern)

      list_of_changes_cols.append(col)
df_4
```

```python
# new data frame with only columns that has changed
df_changes = df_4[list_of_changes_cols].copy().astype(float)
with open('proj2_ex04.pkl', 'wb') as file:
  df_changes.to_pickle(file)
df_changes
```

### Exercise 5: One-hot encoding

```python
pattern = r"^[a-z ]+$"

df_5 = df.copy()
# For each column we check if all 3 requirements are met
# 1) value has to match pattern (only text lower letters)
# 2) column must have less than 10 unique values
# 3) values cannot be in txt file
i = 1
for col in df_5:
  if(df_5[col].astype(str).str.match(pattern).all() and len(df_5[col].unique()) < 10 and all(str(val) not in string_to_val.keys() for val in df_5[col])):
    pd.get_dummies(df_5[col]).to_pickle(f'proj2_ex05_{i}.pkl')
    i += 1
```
